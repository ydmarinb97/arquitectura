```python
import json
os.chdir('C:/Users/ydmar/Documents/Mestria/Analitica de grandes datos/careerbuilder-job-listing-2020')
```

```python
path = 'marketing_sample_for_careerbuilder_usa-careerbuilder_job_listing__20200401_20200630__30k_data.ldjson'
jobs = pd.DataFrame()
with open(path, 'r', encoding="utf8") as f:
    for i, line in enumerate(f):
        try:
            data = json.loads(line.strip('\n'))
            temp = pd.DataFrame([data], columns = data.keys())
            jobs = pd.concat([jobs, temp], axis = 0)
        except Exception as e:
            print(e)
```

```python
jobs.to_csv('trabajos.csv')
```

# Ciudad

```python
df_ciudad = jobs[['city', 'state', 'country']].drop_duplicates()
df_ciudad['city'] = df_ciudad['city'].apply(lambda x: x.upper())
df_ciudad.to_csv('ciudad.csv')
```

```python
df_ciudad['id_ciudad'] = df_ciudad.index
```

```python
df_ciudad
```

## Empresa

```python
df_empresa = jobs[['company_name', 'company_description', 'city']].drop_duplicates()
df_empresa['city'] = df_empresa['city'].apply(lambda x: x.upper())
df_empresa['company_name'] = df_empresa['company_name'].apply(lambda x: x.upper())
```

```python
df_empresa.set_index('city', inplace=True)
df_ciudad.set_index('city', inplace=True)
```

```python
df_empresa = df_empresa.join(df_ciudad['id_ciudad'])
```

```python
df_empresa.reset_index(inplace=True)

del df_empresa['city']
```

```python
df_empresa.to_csv('empresa.csv')
```
